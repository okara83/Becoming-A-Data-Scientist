```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["feature1", "feature2"]].values
```

```python
y=df["price"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_train
```

## SCALING DATA

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler2 = MinMaxScaler()
```

```python
scaler2.fit(X_train)
```

```python
X_train = scaler2.transform(X_train)
```

```python
X_test = scaler2.transform(X_test)
```

```python
X_train.min()
```

```python
X_train.max()
```

TensorFlow Syntax:

(use %tensorflow_version 1.x to use old versions of tf)

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Activation
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss", verbose=1, patience=15)
```

```python
from tensorflow.keras.optimizers import Adam, RMSprop
```

```python
model6 = Sequential()
model6.add(Dense(4, input_dim=2, activation="relu"))           #input_dim = 2  or input_dim = X_train.shape[0]
model6.add(Dense(4, activation="relu"))
model6.add(Dense(2, activation="relu"))
model6.add(Dense(1))
opt = Adam(learning_rate=0.001)
model6.compile(optimizer=opt, loss="mse")
```

```python
model6.fit(X_train, y_train, validation_data = (X_test, y_test), batch_size=32, epochs=750, callbacks=[early_stop])
```

```python
model6.summary()
```

```python
loss_df=pd.DataFrame(model6.history.history)
```

```python
loss_df.plot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
loss = model6.history.history['loss']
sns.lineplot(x=range(len(loss)),y=loss)
plt.title("Training Loss per Epoch");
```

```python

fig, ax= plt.subplots()
ax.plot(loss_df)
ax.set_xlabel("epoch")
ax.set_ylabel("loss")
ax.set_title("loss - epoch graph")
ax.set_xlim([0,250]) # focusing on the given x values
ax.set_ylim([0,50]) # focusing on the given y values
```

```python
loss_df.loss.idxmin() 
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\n","mae:", mae, "\n","mse:",mse, "\n","rmse:",rmse)
```

```python
y_pred = model6.predict(X_test)
```

```python
eval_metrics(y_test, y_pred)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
my_dict={"Actual":y_test, "Pred":y_pred.reshape(300,)}
compare=pd.DataFrame(my_dict)
```

```python
compare.sample(5)
```
