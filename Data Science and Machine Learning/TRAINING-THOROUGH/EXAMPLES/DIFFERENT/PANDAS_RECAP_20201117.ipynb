## Operations with str

```python
data2=sns.load_dataset("mpg")
```

```python
data2.head(2)
```

### Contains

```python
data2.name.str.contains("chevrolet").count()
```

```python
data2.loc[data2.name.str.contains("chevrolet")].head(3)
```

```python
data2.loc[~data2.name.str.contains("chevrolet")].head()
```

```python
data2.loc[(data2.name.str.contains("chevrolet")) & (data2.origin.str.contains("usa"))].head()
```

```python
data2.loc[data2.name.str.contains("chevrolet|buick", regex=True)].head()
```

```python
data2.loc[data2.name.str.contains("CHEVROLET|BUICK", case=False, regex=True)].head()
```

### Contains with regex

#all name values starting with "bu"

```python
data2.loc[data2.name.str.contains("^bu", case = False, regex = True)].head()
```

#all name values ending with "lo"

```python
data2.loc[data2.name.str.contains('lo$', case=False, regex=True)]
```

### Regex-Extracting

#First 5 characters from name column

```python
data2.name.str.extract(r'(^\w{5})')[160:165]
```

### Using regex-match

```python
data2.name.str.match(r'(^[C|c].*)').count()
```

### Replacing values

```python
data2.cylinders.replace({8:"eight", 6:"six"})[100:105]
```

```python
data2.origin.replace(["japan", "europe"], "others")
```

```python
data2.loc[data2.model_year<80, "Type"] = "Classic"
data2.head()
```
