```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open("dolphin.json","r") as f:
    A = json.load(f)
```

```python
g_dolphin =nx.from_numpy_matrix(np.matrix(A))
g_karate = nx.karate_club_graph()
d = dict(g_dolphin.degree)


plt.figure()
plt.hist([x[1] for x in list(g_dolphin.degree)])
plt.show()



plt.figure()
nx.draw_kamada_kawai(g_dolphin,  node_size=[v * 100 for v in d.values()])
plt.show()

pos = nx.spring_layout(g_dolphin)


plt.figure()
nx.draw(g_dolphin, pos=pos,  node_color=[v * 100 for v in d.values()],cmap='jet')
plt.show()
```

```python
Bet=nx.betweenness_centrality(g_dolphin)
Clo=nx.closeness_centrality(g_dolphin)
Deg=nx.degree_centrality(g_dolphin)
Eig = nx.eigenvector_centrality(g_dolphin)



plt.figure()
plt.plot([Bet[x] for x in g_dolphin.nodes()], [Eig[x] for x in g_dolphin.nodes()],'o',color='orange')
plt.xlabel('Betweenness')
plt.ylabel('Eigenvector')
plt.show()

plt.figure()
plt.plot([Deg[x] for x in g_dolphin.nodes()], [Clo[x] for x in g_dolphin.nodes()],'o',color='blue')
plt.xlabel('Degree')
plt.ylabel('Closeness')
plt.show()

```
