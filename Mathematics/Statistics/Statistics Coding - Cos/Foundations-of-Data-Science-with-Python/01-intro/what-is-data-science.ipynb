# What is data science? 

Our world is filled with information. In fact, many times the amount of information we have access to can be overwhelming. We start by considering when information becomes *data*:

````{card}
DEFINITION
^^^
```{glossary}
data
     Collections of measurements, characteristics, or facts about a group.
```
````




Data consists of *variables*:


````{card}
DEFINITION
^^^
```{glossary}
variable
     A particular characteristic, measurement, or fact about a particular member of a group.
```
````



Variables maybe either *quantitative* or *qualitative*:

````{card}
DEFINITION
^^^
```{glossary}
quantitative data
    Numeric data. Quantitative data may be either discrete (such as the number of people in a family) or continuous (such as grade point average).
```
````






````{card}
DEFINITION
^^^
```{glossary}
qualitative data
    Non-numeric data. Qualitative variables are generally non-numeric categories that data may belong to (such as hair color). Some categories may have an order associated with them, but the order does not imply a numeric nature to the categories; for example, cancer may be classified as Stage 0 through Stage IV, with each stage representing increasing spreading of the cancer.  ```
````






**Examples of quantitative variables:**
- height
- weight
- yearly income
- college GPA
- miles driven commuting to work

**Examples of qualitative variables:**
- hair color (blond, brown, black, red, gray, ...)
- current precipitation status (clear, sprinkling, raining, pouring)
- car type (sedan, couple, SUV, minivan, ...)
- stage of cancer in a cancer patient (0, I, II, III, IV)




As engineers, our goal is to make sense of the world and use what we learn to take action. **The general goal of data science is to **generate meaning from data.** In particular, data science applies computational tools and mathematical methods to transform data and extract signficant features from it.

Data scientists often start with a research question. For instance:
* How fast was the COVID-19 corona virus spreading when it first became prevalent in the United States in the Spring of 2020?
* Do gun laws affect firearms mortality?
* Is the climate changing?
* How reliable are mammograms (x-rays used to detect breast cancer)?

<!-- * **ADD MORE QUESTIONS FROM BOOK HERE** -->

One of the goals of a data scientist is to take broad research questions and translate them into questions that can be answered using data. For instance, here are some ways that the questions above may be reformulated so that they can be answered with data:
* Instead of "Is the climate changing?", a data scientist may ask a much more specific question, such as "Has the temperature in Miami, FL increased over the past 40 years?"
* Instead of "Do gun laws affect firearms mortality?", a data scientist may ask about a specific law, such as "How did the federal assault weapons ban of XXXX affect firearms mortality?" Or the data scientist may look at the affect of different numbers of laws across states: "How does the number of state gun laws affect firearms mortality?" 
* To determine how reliable mammograms are, a data scientist may ask "If a patient's mammogram comes back positive, what is the probability that patient has breast cancer?"
* In assessing the rate of spread of the COVID-19 virus, the data scientst may ask "Was the number of cases growing exponentially in March of 2019? If so, what was the exponential growth rate?"
