# Calculating $p$-Values

The simplest and most common test for statistical significance is to estimate a $p$-value from the data, where $p$ is the probability of a random sample from $\mathcal{P}$ having a test difference  "at least as extreme" as the observed difference in summary statistics, $\Delta$, **under the null hypothesis**. Let $E$ be the event that an arbitrary sample test difference $\Delta_i$  is "at least as extreme" as $\Delta$. Then we use the notation $P(E|H_0)$ to denote the probability that $E$ occurs **under the null hypothesis**. (The vertical bar "|" denotes a conditional probability, and we delve more into the meaning of these types of probabilities in {doc}`Chapter 6<../06-conditional-prob/intro>`.)


To estimate $p$ using resampling, we use a simulation of the form described in the last section. We estimate $P(E|H_0)$ by using the relative frequency 

$$
r_E(N) = \frac{ n_E(N)}{N},
$$
where $n_E(N)$ is the number of times $E$ is observed in $N$ trials. The only remaining issue to resolve is how to determine whether $E$ occurred based on the simulated test differences, $\Delta_i$.

## One-Tail or Two-Tail Tests

NHSTs are classified as one-tailed or two-tailed (also called one-sided or two-sided) based on the way that "at least as extreme" as $\Delta$ is interpreted.  Consider again our  statistical questions from the last section:
1. Is the observed difference in median wealth ($\Delta = 57,400$) between the undergrad and grad groups statistically significant?
2. Is the observed difference in the relative frequency of millionaires between the undergrad and grad groups ($\Delta \approx 0.053$) statistically significant?



Let's use the first one of these as an example to understand one-tailed versus two-tailed tests, let $\tilde{U}_i$ and $\tilde{G}_i$ denote the **observed median values** for the $i$th undergrad and grad samples. Recall that we previously defined $\Delta_i = \tilde{G}_i - \tilde{U}_i$.


In many statistical tests, under the null hypothesis, the difference between summary statistics will result in values that are equally likely to be positive and negative, with most of the values centered around 0. (In Chapters 8 and 9, we will introduce formal terminology for this: the difference has zero mean and its distribution is symmetric around the mean.)

Below is an example that shows the type of histrogram that we expect to encounter.

`````python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
T=stats.t(4)
tvals=T.rvs(100_000)
plt.hist(tvals, bins =np.arange(-4,4.5,0.5), alpha=1
        );
plt.xlabel('Difference Values')
plt.ylabel('Number of Occurrences (out of 10,000)');
`````

In this histogram, the values with the higher counts correspond to those with higher probability of occurring. The way the probabilities map to the values is called the *distribution* of the random values. (Again, this is discussed in more detail in Chapter 8.) We can define the *mode* for a distribution of random values:


````{card}
DEFINITION
^^^
```{glossary}
mode(s) of a distribution    
    For a distribution of random values,  the value(s) with the maximum probability (or probability density -- see Chapter 8) are the *modes* of the distribution.
```
````


We can also classify the type of distribution based on how many modes it has. The most common case, and the main one we will utilize in this book is when a distribution has a single mode:

````{card}
DEFINITION
^^^
```{glossary}
unimodal distribution    
   A distribution of random values that has a single mode.
```
````



Because the total probability in the distribution must be 1, most unimodal distributions  have the probability decrease as the values get farther from the mode of the distribution. We can then define a *tail probability*:


````{card}
DEFINITION
^^^
```{glossary}
tail probability
    For a unimodal distribution in which the probabilities decrease as some function of the distance from the mode, a *tail probability* is the probability of being at least some distance from the mode.
```
````


There are two types of tail probabilities: one-sided tails and two-sided tails. Let $M_\Delta$ be the mode of the distribution of the test difference. One-sided tails can either be:


````{card}
DEFINITION
^^^
```{glossary}
right, or upper, tail
    For a random value $\Delta$ that has a distribution with mode $M_\Delta$, the *right tail* or *upper tail* is of the form $P(\Delta > M_\Delta +d)$  for some value $d>0$..
```
````



````{card}
DEFINITION
^^^
```{glossary}
left, or lower, tail
    For a random value $\Delta$ that has a distribution with mode $M_\Delta$, the *left tail* or *lower tail* is of the form $P(\Delta < M_\Delta -d)$ for some value $d>0$.
```
````


These two types of tails are shown in the diagrams below:

`````python
T=stats.t(4)
tvals=T.rvs(100_000)
fig, ax = plt.subplots(2, figsize=(6,7))

for i in range(2):
  ax[i].hist(tvals, bins =np.arange(-4,4.5,0.5), alpha=0.5 );
  ax[i].set_xlabel('Difference Values, $\Delta_i$')
  ax[i].set_ylabel('Counts (out of 10,000)');
  
thresh=1.5
greater = tvals[tvals>thresh]
ax[0].hist(greater, bins =np.arange(thresh,4.5,0.5), alpha=1 );
ax[0].annotate("$P(\Delta > M_\Delta+1.5)$",
            xy=(2.3,2500), xycoords='data',
            xytext=(2.5, 10_000), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
ax[0].set_title('Right, or upper, tail')



lesser = tvals[tvals<-thresh]
ax[1].hist(lesser, bins =np.arange(-4,-thresh+0.5,0.5), alpha=1 );
ax[1].annotate("$P(\Delta < M_\Delta-1.5)$",
            xy=(-2.3,2500), xycoords='data',
            xytext=(-3.7, 10_000), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
ax[1].set_title('Left, or lower, tail')


fig.tight_layout()
plt.subplots_adjust(bottom=0.1,  
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
`````

A two-sided tail simply includes the probability of both upper and lower tails:

````{card}
DEFINITION
^^^
```{glossary}
two-sided tail
    For a random value $\Delta$ that has a distribution with mode $M_\Delta$, a *two-sided tail* is of the form $P(|\Delta - M_\Delta| > d)$ for some value $d>0$.
```
````

A two-sided tail is the union of a left tail and a right tail. The following figure illustrates a two-sided tail:

`````python
T=stats.t(4)
tvals=T.rvs(100_000)
fig, ax = plt.subplots(1, figsize=(6,7))

for i in range(1):
  ax.hist(tvals, bins =np.arange(-4,4.5,0.5), alpha=0.5 );
  ax.set_xlabel('Difference Values, $\Delta_i$')
  ax.set_ylabel('Counts (out of 10,000)');
  
thresh=1.5
greater = tvals[tvals>thresh]
ax.hist(greater, bins =np.arange(thresh,4.5,0.5), alpha=1 ,
        color='C1' );
ax.annotate("$P( |\Delta - M_\Delta | > 1.5 )$",
            xy=(2.3,2500), xycoords='data',
            xytext=(2.5, 10_000), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
ax.set_title('Right, or upper, tail')



lesser = tvals[tvals<-thresh]
ax.hist(lesser, bins =np.arange(-4,-thresh+0.5,0.5), alpha=1,
        color='C1' );
ax.annotate('',
            xy=(-2.3,2500), xycoords='data',
            xytext=(3.4, 9_700), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
ax.set_title('Left, or lower, tail')


fig.tight_layout()
plt.subplots_adjust(bottom=0.1,  
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
`````

### One-tailed Test

A one-tailed test is **concerned about the particular direction of the observed difference** in test statistics between the two groups. A one-tailed test on median wealth estimates the probability of the grad group's median wealth at least $\$57,400$ higher than the undergrad group's median wealth, under $H_0$. For a given test difference $\Delta_i$, we say that $E$ occurs if $\Delta_i \ge 57,400$. Thus the one-tailed test estimate $P( \Delta_i > 57,400 |H_0)$. 




If we collected all the values of $\Delta_i$ in our simulation into a NumPy array `diffs`, then we can calculate the relative frequency of the observed difference `Delta_i` exceeding a threshold `thresh`,  by
* comparing the array to the threshold `diffs >= thresh`, and
* counting the number of `True` values; since `False` is counted as zero, we can do `np.count_nonzero(diffs >= thresh)`
* dividing the number of values where `diffs >=thresh` by the number of simulated test differences





Because I am not ready to discuss how to draw samples from the pooled data yet in this chapter, let's illustrate this idea by generating an array of 100 random values in the interval $[-0.5, 0.5)$ to represent an array of simulated differences. We estimate the probability that the simulated differences are bigger than 0.9. Run this code a few times and observe the outcomes. Increase the number of simulated differences to 100,000. What do you observe?

`````python
import numpy as np
import numpy.random as npr

num_sims=100
thresh = 0.4

diffs = npr.rand(num_sims) - 0.5 # Random values on [-0.5, 0.5)
np.count_nonzero( diffs >= thresh) / num_sims
`````

Note that if we only wish to estimate the $p$-value, then there is no need to store all the test differences. We can instead initialize a counter to 0 at the beginning of the simulation and then increment the counter for each test difference that exceeds the threshold.

### Two-tailed Test

A two-tailed test is only concerned that the test statistics of the two groups differ by at least as much as the observed difference $\Delta$; it is **not concerned with the direction of the difference**. A two-sided test on median wealth estimates the probability that the difference in wealth between the two groups is at least $\$57,400$. Unpacking the last sentence a bit, $E$ occurs if *either* $ \tilde{G}_i - \tilde{U}_i \ge 57,400$ **or** $ \tilde{U}_i - \tilde{G}_i \ge 57,400$. These two cases are mutually exclusive, and we can write the combined event in more concise notation as $|\tilde{G}_i - \tilde{U}_i | \ge 57,400$, which is the same as $|\Delta_i|  \ge 57,400$, where the vertical bars indicate absolute value. Thus, the two sided test estimates 

$$
P\bigl(|\Delta_i| > 57,400 \bigl \vert H_0 \bigr .\bigr).
$$ 
In this last expression, the first pair of vertical bars indicate absolute value, whereas the last vertical bar indicates that we are calculating this probability under the null hypothesis, $H_0$.



Given an array `diffs` of all the values of $\Delta_i$ in our simulation, we can calculate the relative frequency of the absolute value of the difference `Delta_i` exceeding a threshold `thresh`  by
* comparing the absolute value of the array to the threshold `np.abs(diffs) >= thresh`, and
* counting the number of `True` values; since `False` is counted as zero, we can do `np.count_nonzero( np.abs(diffs) >= thresh)`
* dividing the number of values where `np.abs(diffs) >=thresh` by the number of simulated test differences

The Python code is just a minor modification of that for the one-sided test: we just take the *absolute value* of the test differences before comparing them to the threshold:

`````python
import numpy as np
import numpy.random as npr

num_sims=1000
thresh = 0.4

diffs = npr.rand(num_sims) - 0.5
np.count_nonzero( np.abs(diffs) >= thresh) / num_sims
`````

Note that the $p$-value for the two-sided test is always higher than for the one-sided test because the two-sided probability  is the union of two one-sided probabilities.

## Choosing Between a One-Sided or Two-Sided Test



The choice of whether to use a one-sided or two-sided test depends primarily on the initial research question. If the initial research question specifies a particular direction, then a one-sided test is justified. For instance, if the initial research question is: does post-baccalaureate education increase mediate net family wealth?, then a one-sided test is justified. However, if the initial research question asks only asks about a difference in effects, then a two-sided test should be used. For instance, if the initial research question is: is there a difference between those with an undergraduate education  and those with post-baccalaureate education, then a two-sided test should be used. Because a two-sided test produces higher $p$-values, it is less likely to be smaller than the threshold, $\alpha$. Thus, a two-sided test is a more conservative test -- it is less likely to find statistical significance than a one-sided test. When in doubt, it is best to choose a two-sided test.
