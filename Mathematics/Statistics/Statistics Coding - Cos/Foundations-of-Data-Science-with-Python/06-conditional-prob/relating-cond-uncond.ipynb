To see how conditional information can change probabilities, let's consider a few examples:

**1.** Suppose a **fair** coin is flipped twice, and let $H_i$ denote the event that the coin came up heads on flip $i.$ Then clearly, $P(H_i) =1/2$ and $P(H_1 \cap H_2) = 1/4$. From the definition of conditional probability,

$$
P(H_2|H_1) = \frac{P(H_1 \cap H_2)}{P(H_1)} = \frac{1/4}{1/2} = 1/2
$$

So in this case, $P(H_2|H_1)= P(H_2)$ and **conditional information did not change the probability**. This is expected because the outcome of the second flip should be independent of what happened on the first flip.

**2.** Suppose I have a box containing two balls, one of which is white and one of which is black. Balls are drawn consecutively from the box **without replacement**. Let $W_i$ be the event that the $i$th draw is a white ball.  If we are told that the first draw was a white ball, then clearly the remaining ball is black, so $P(W_2|W_1)=0$. Let's check using the definition of conditional probability. From our previous examples, we know that $P(W_i)=1/2$, and $P(W_1 \cap W_2) = 0$ because there is no way to draw white balls out on both draws.  Thus,

$$
P(W_2 |W_1) = \frac{P(W_1 \cap W_2)}{P(W_1)} = \frac{0}{1/2} = 0.
$$

So, in this case $P(W_2|W_1)< P(W_2)$, and **conditional information reduced the probability**.

In fact, this will happen whenever the two events being evaluated are mutually exclusive. Let $A$ and $B$ be mutually exclusive events with $P(A)>0$ and $P(B) >0$. Then

$$
P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{P(\emptyset)}{P(B)} = 0.
$$

If two events are mutually exclusive, then if we know one of them occurs, then the other **cannot have occurred** because it shares no outcomes in common with the event that occurred.

**3.** Consider again the scenario where a fair coin is flipped two times and the ordered result is noted. Let $E_i$ be the result that the number of heads observed is $i$, and let $F$ be the result that at least one heads occurred. 

From counting, we see $P(E_1) = 1/2$ and $P(F) = 3/4$. Let's find $P(F|E_1)$ using the definition of conditional probability. We need to calculate $P(F \cap E_1)$. We can write out the relevant events in terms of their outcomes as $E_1=\{ (H,T), ~(T,H) \}$ and $F= \{ (H,T),~,(T,H),~,(H,H) \}$. Then $F \cap E_1 = E_1$.  So,

$$
P(F|E_1) = \frac{ P(F \cap E_1)}{P(E_1)} = \frac{ P(E_1)}{P(E_1)} = 1.
$$

So, in this case, $P(F|E_1) > P(F)$, and **conditional information increased the probability**.

In general, if we have events $A$ and $B$ where $B$ is a subset of $A$, then $A \cap B = B$. Then if 
$P(A)<1$ and $P(B)>0$,
$$
P(A|B) =\frac{P(A \cap B)}{P(B)} = \frac{P(B)}{P(B)} = 1 > P(A).
$$


**In conclusion, there is no general answer to how conditional information will change the probability of an event. The conditional probability may be greater than, less than, or equal to the unconditional probability, depending on the relation between the two events.**
