```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt




plt.figure(1, figsize=(5, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

x = np.linspace(.001, 1.5, 1000)
y = .25/x

xx = np.linspace(.001, 1., 1000)
yy = .25/xx

plt.rc('font', family='serif')
plt.plot(x, y, color='black', linewidth=1)
plt.plot([0, 1], [1, 1], color='black', linewidth=1, linestyle='--')
plt.plot([1, 1], [0, 1], color='black', linewidth=1, linestyle='--')
plt.ylim((0, 1.5))
plt.xlim((0, 1.5))
plt.xticks([0, 1], size=15)
plt.yticks([0, 1], size=15)
plt.fill_between(xx, 0, np.minimum(yy, 1), facecolor='grey', alpha=.5)
plt.ylabel(r'$y$', size=20)
plt.xlabel(r'$x$', size=20)



plt.annotate(r'$ y= \left(\frac{1}{4}\right)\frac{1}{x}$', size=20, xy=(.5, .22),  xycoords='data',
            xytext=(.5, 0.55), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top', rotation=-35
            )





plt.tight_layout()
plt.savefig('prob19.pdf', format='pdf');
```

```python
plt.figure(1, figsize=(5, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

x = np.linspace(0, 1, 1000)
xx = np.linspace(.5, 1, 1000)
xxx = np.linspace(0, .5, 1000)

yl=x*0
yu=x*0+1


plt.rc('font', family='serif')
plt.plot(x, x, color='black', linewidth=1)
plt.plot(xx, xx-.5, color='black', linewidth=1)
plt.plot(xxx, xxx+.5, color='black', linewidth=1)
plt.plot([0, 1], [1, 1], color='black', linewidth=1, linestyle='--')
plt.plot([1, 1], [0, 1], color='black', linewidth=1, linestyle='--')
plt.ylim((0, 1.5))
plt.xlim((0, 1.5))
plt.xticks([0, 1], size=15)
plt.yticks([0, 1], size=15)
plt.fill_between(x, np.maximum(yl, x-.5), np.minimum(yu, x+.5), alpha=0, hatch='\\')
plt.fill_between(x, x, 1, facecolor='black', alpha=.2)
plt.ylabel(r'$y$', size=20)
plt.xlabel(r'$x$', size=20)

plt.annotate(r'$y = x + 0.5$', xy=(1.1, .22),  xycoords='data',
            xytext=(.55, 0.96), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top',rotation=45, size=15
            )

plt.annotate(r'$y = x-0.5$', xy=(1.1, .22),  xycoords='data',
            xytext=(.9, 0.6), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top',rotation=45, size=15
            )


plt.annotate(r'$y = x $', xy=(1.1, .22),  xycoords='data',
            xytext=(.81, 0.825), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top',rotation=45, size=15
            )



plt.tight_layout()
plt.savefig('prob30.pdf', format='pdf');
```
